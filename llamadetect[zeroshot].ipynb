```python


# import packages
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split


data = pd.read_csv('/content/All(4430R).csv')

y_train, y_test = train_test_split(
  data , random_state=104,test_size=0.20, shuffle=True)
```

```python
import torch
import pandas as pd
from datasets import Dataset
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
import torch
import pandas as pd
from datasets import Dataset
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline
from sklearn.metrics import classification_report, PrecisionRecallDisplay, accuracy_score, f1_score, precision_score, recall_score

def evaluate_zero_shot_fake_news(model_name="meta-llama/Llama-3.2-1B", test_file=y_test):
    """
    Evaluates a pre-trained Llama model for zero-shot fake news classification.

    Parameters:
    - model_name: The pre-trained model to use.
    - test_file: Path to the test dataset.

    Returns:
    - Prints classification metrics and displays Precision-Recall curve.
    """

    # Load model and tokenizer
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    model = AutoModelForSequenceClassification.from_pretrained(model_name)

    # Move model to GPU if available
    device = "cuda" if torch.cuda.is_available() else "cpu"
    model.to(device)

    # Define a zero-shot classification pipeline
    pipe = pipeline("text-classification", model=model, tokenizer=tokenizer, device=0 if device == "cuda" else -1)

    # Load test dataset
    test_df =y_test

    # Convert to Hugging Face dataset
    test_data = Dataset.from_pandas(test_df)

    # Function to get predictions
    def label_score(texts):
        predictions = pipe(texts)
        return [1 if pred["label"] == "LABEL_1" else 0 for pred in predictions]  # Assuming "LABEL_1" means fake news

    # Get true labels and predictions
    probas = test_df["Response"].apply(lambda x: label_score([x])[0])
    preds = probas.apply(lambda x: 1 if x > 0.5 else 0)  # Convert scores into binary predictions

    # Compute evaluation metrics
    report = classification_report(test_df["Label"], preds)
    print(report)


# Run evaluation
evaluate_zero_shot_fake_news()
```
